```python
# load statmodels functions
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# compute the vif for all given features
def compute_vif(df,considered_features):
    
    X = df[considered_features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
```

```python
# features to consider removing
considered_features = ['g', 'u', 'r', 'i', 'z', 'plate', 'specobjid','mjd']

# compute vif 
compute_vif(data_explore,considered_features).sort_values('VIF', ascending=False)
```

#Anlisis discriminante Cuadrtico 

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
def QDA(data_set, folds, priors):
  X = data_set.drop('class',axis=1)
  Y = data_set['class']
  resultados = pd.DataFrame()
  idx = 0
  for prior in priors:
    for fold in folds:
      skf = StratifiedKFold(n_splits=fold)
      scores = []
      for train, test in skf.split(X, Y):
        Xtrain, Xtest = X.iloc[train], X.iloc[test]
        Ytrain, Ytest = Y.iloc[train], Y.iloc[test]
        modelo = QuadraticDiscriminantAnalysis(priors=prior)
        modelo.fit(X=Xtrain,y=Ytrain)
        yPred = modelo.predict(X=Xtest)
        score = balanced_accuracy_score(y_true=Ytest, y_pred=yPred)
        scores.append(score)
      resultados.loc[idx,'score']=np.mean(scores)
      resultados.loc[idx,'priors']=prior
      resultados.loc[idx,'folds']=fold
      idx=idx+1
  return resultados
```

```python
qda_outliers_results = QDA(data_set=data_frame,folds=[4,5,6], priors=[1,0.1,0.01,0.001])
qda_outliers_results.sort_values(['score'],ascending=[False]).head(5)
qda_outliers_results.to_csv('qda.csv')
```

```python
qda_no_outliers_results = QDA(data_set=no_outliers_data,folds=[4,5,6], priors=[1,0.1,0.01,0.001])
qda_no_outliers_results.sort_values(['score'],ascending=[False]).head(5)
qda_no_outliers_results.to_csv('qda_no_outliers.csv')
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def KNN(data_set,neighbors,folds):
  X = data_set.drop('class',axis=1)
  Y = data_set['class']
  resultados = pd.DataFrame()
  idx = 0
  for n in neighbors:
    for fold in folds:
      skf = StratifiedKFold(n_splits=fold)
      scores = []
      for train, test in skf.split(X, Y):
        Xtrain, Xtest = X.iloc[train], X.iloc[test]
        Ytrain, Ytest = Y.iloc[train], Y.iloc[test]
        modelo = KNeighborsClassifier(n_neighbors=n)
        modelo.fit(X=Xtrain,y=Ytrain)
        yPred = modelo.predict(X=Xtest)
        score = balanced_accuracy_score(y_true=Ytest, y_pred=yPred)
        scores.append(score)
      resultados.loc[idx,'score']=np.mean(scores)
      resultados.loc[idx,'n_neighbors']=n
      resultados.loc[idx,'folds']=fold
      idx=idx+1
  return resultados
```

```python
knn_outliers_results = KNN(data_set=data_frame,folds=[4,5,6], neighbors=[3,5,7,10])
knn_outliers_results.to_csv('knn.csv')
knn_outliers_results
```

```python
knn_no_outliers_results = KNN(data_set=no_outliers_data,folds=[4,5,6], neighbors=[3,5,7,10])
knn_no_outliers_results.to_csv('knn_no_outliers.csv')
knn_no_outliers_results
```

#Gradiente Boosting Tree 

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
def GBT(data_set,folds,trees,split_samples):
  X = data_set.drop('class',axis=1)
  Y = data_set['class']
  resultados = pd.DataFrame()
  idx = 0
  for tree in trees:
    for samples in split_samples:
      for fold in folds:
        skf = StratifiedKFold(n_splits=fold)
        scores = []
        for train, test in skf.split(X, Y):
          Xtrain, Xtest = X.iloc[train], X.iloc[test]
          Ytrain, Ytest = Y.iloc[train], Y.iloc[test]
          modelo = GradientBoostingClassifier(min_samples_split=samples,n_estimators=tree)
          modelo.fit(X=Xtrain,y=Ytrain)
          yPred = modelo.predict(X=Xtest)
          score = balanced_accuracy_score(y_true=Ytest, y_pred=yPred)
          scores.append(score)
        resultados.loc[idx,'score']=np.mean(scores)
        resultados.loc[idx,'trees']=tree
        resultados.loc[idx,'split_sample']=samples
        resultados.loc[idx,'folds']=fold
        idx=idx+1
  return resultados
```

```python
gbt_outliers_results = GBT(data_set=data_frame,folds=[4,5,6], trees=[3,5,7,10], split_samples=[2,3,5])
gbt_outliers_results.to_csv('gbt.csv')
gbt_outliers_results.sort_values(['score'],ascending=[False]).head(5)
```

```python
gbt_no_outliers_results = GBT(data_set=no_outliers_data,folds=[4,5,6], trees=[3,5,7,10], split_samples=[2,3,5])
gbt_no_outliers_results.to_csv('gbt_no_outliers.csv')
gbt_no_outliers_results.sort_values(['score'],ascending=[False]).head(5)
```

#Mquinas de Soporte Vectorial

```python
from sklearn.svm import SVC
import itertools
import multiprocessing
from sklearn.multiclass import OneVsRestClassifier
```

```python
def SVM(data_set,folds,kernels, gammas,params_reg):
  X = data_set.drop('class',axis=1)
  Y = data_set['class']
  cores=multiprocessing.cpu_count()-2
  kernels_gammas_regs = list(itertools.product(kernels, gammas, params_reg))
  resultados = pd.DataFrame()
  idx = 0
  for params in kernels_gammas_regs:
    kernel, gamma, param_reg = params
    for fold in folds:
      skf = StratifiedKFold(n_splits=fold)
      scores = []
      for train, test in skf.split(X, Y):
        Xtrain, Xtest = X.iloc[train], X.iloc[test]
        Ytrain, Ytest = Y.iloc[train], Y.iloc[test]
        modelo = OneVsRestClassifier(SVC(kernel=kernel,gamma=gamma,C=param_reg),n_jobs=cores)
        #modelo = SVC(kernel=kernel,gamma=gamma,C=param_reg)
        modelo.fit(X=Xtrain,y=Ytrain)
        yPred = modelo.predict(X=Xtest)
        score = balanced_accuracy_score(y_true=Ytest, y_pred=yPred)
        scores.append(score)
      resultados.loc[idx,'score']=np.mean(scores)
      resultados.loc[idx,'kernel']=kernel
      resultados.loc[idx,'gamma']=gamma
      resultados.loc[idx,'C']=param_reg
      resultados.loc[idx,'folds']=fold
      idx=idx+1
  return resultados
```

```python
svm_outliers_results = SVM(data_set=data_frame,folds=[4,5,6], kernels=['linear', 'rbf'], gammas = [0.01,0.1], params_reg = [0.001, 0.01,0.1, 1.0,10],)
svm_outliers_results.to_csv('svm.csv')
svm_outliers_results.sort_values(['score'],ascending=[False]).head(5)
```
